```python
# Display average customer amount invoiced per year, month, or day
invoice_data_by_customer.groupby(["InvoiceDate_MonthName"]).sum().reindex(constants.month_names).plot.pie(x="InvoiceDate_MonthName", y="InvoiceAmount", title="Total Invoice Amount ($) by Month (2015 - 2020)", figsize=(8, 8))

invoice_data_by_customer.groupby(["InvoiceDate_MonthName"]).mean().reindex(constants.month_names).plot.pie(x="InvoiceDate_MonthName", y="InvoiceAmount", title="Average Invoice Amount ($) by Month (2015 - 2020)", figsize=(8, 8))

invoice_data_by_customer.groupby(["InvoiceDate_MonthName"]).count().reindex(constants.month_names).plot.pie(x="InvoiceDate_MonthName", y="InvoiceAmount", title="Number of Invoices per Month (2015 - 2020)", figsize=(8, 8))
```
